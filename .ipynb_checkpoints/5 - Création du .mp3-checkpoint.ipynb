# 5 - Création du .mp3

**Input du notebook** : 
- Une liste des notes détectées dans les précédents notebooks de la forme d'une liste ['Mi', 'La', 'Do_aigu', ...]

**Output du notebook** : 
- Un fichier .mp3 contenant la musique de la portée.

```python
from preprocessing import *
# moviepy va permettre la complilation de notes individuelles en un morceau de musique
# http://zulko.github.io/moviepy/getting_started/audioclips.html
from moviepy.editor import * 
import pickle 
import glob
```

On enregistre les images de notes dans le dossier `A_predire` avec la ligne suivante : 

```python
prédictions    = pickle.load(open('predictions.p', 'rb'))
partition_name = 'bellaciao.png'
```

On peut afficher la partition détectée : 

```python
print(len(prédictions), 'note détectées.')
plt.figure(figsize=(15,5))
img = mpimg.imread("Partitions/"+partition_name)
plt.imshow(img)
plt.axis('off')
plt.show()

for note in prédictions:
    print(note, end=' ')
```

Et la lire :

```python
mapping = {'Do' :'Piano.ff.A3.aiff',
           'Re' :'Piano.ff.B3.aiff',
           'Mi' :'Piano.ff.C4.aiff',
           'Fa' :'Piano.ff.D4.aiff',
           'Sol':'Piano.ff.E4.aiff',
           'La' :'Piano.ff.F4.aiff',
           'Si' :'Piano.ff.G4.aiff',
           'Do_aigu' :'Piano.ff.A4.aiff',
           'Re_aigu' :'Piano.ff.B4.aiff',
           'Mi_aigu' :'Piano.ff.C5.aiff', 
           'Fa_aigu' :'Piano.ff.D5.aiff',
           'Sol_aigu':'Piano.ff.E5.aiff',
           'La_aigu' :'Piano.ff.F5.aiff',
           'Si_aigu' :'Piano.ff.G5.aiff',}
```

```python
lets_play = []
for note in prédictions:
    audioclip = AudioFileClip("Notes_sons/" + mapping[note]).set_duration(1)
    lets_play.append(audioclip)
```

```python
lets_play = concatenate_audioclips(lets_play)
lets_play.write_audiofile('lets_play.mp3' )
```

Lecture des notes détectées sur la partition : 

```python
from IPython.display import Audio
Audio('lets_play.mp3', autoplay=True, )
```
