# 1 - Constitution du Train-set

**Input du notebook** : 
- une partition au format png et d'une qualité suffisante (dossier `Partitions/`)

**Output du notebook** : 
- dans le dossier `temp`, l'image de la partition disséquée en morceaux de taille LARG x LONG (14x72 px).

```python
import pandas as pd
import os
import numpy as np ; np.random.seed(123)
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import glob
from skimage.transform import resize
```

## Chargement de l'image et conversion en noir et blanc

```python
img = mpimg.imread("Partitions/g_du_tabac.png")
print(img.shape)
img = img[:,:,:min(img.shape[2], 3)]
print(img.shape)
```

```python
plt.figure(figsize=(20,10))
plt.imshow(img)
plt.title('Image initiale', size=18)
plt.show()
```

Peu importe la couleur d'une note de musique. On peut donc convertir la partition en noir et blanc : 

```python
if len(img.shape) > 2:
    img_noir_blanc = list(map(lambda x : [z>0.8 for y in x for z in y], img))
    img_noir_blanc = np.array(img_noir_blanc).reshape((img.shape[0], img.shape[1], 3))
    img_noir_blanc = np.array([[y[0] for y in x] for x in img_noir_blanc]).reshape((img.shape[0], img.shape[1]))
else :
    img_noir_blanc = img
print("L'image fait désormais", img_noir_blanc.shape[0], "pixels de large sur", img_noir_blanc.shape[1], "de long.")
```

```python
plt.figure(figsize=(20,10))
plt.imshow(img_noir_blanc)
plt.title('Image en noir et blanc', size=18)
plt.show()
```

### Détection de la portée

Commençons par trouver les 5 lignes de portée. On teste 3 épaisseurs de trait, avec à chaque fois le pourcentage de pixels noirs que l'on s'attend à trouver sur la ligne. 

```python
for seuil, épaisseur_portée in [(0.1, 1), (0.45, 2), (0.66, 3)]:
    abscisses = []
    for i in np.arange(0, img_noir_blanc.shape[0], épaisseur_portée):
        if np.sum(img_noir_blanc[i,:]) < seuil*img_noir_blanc.shape[1]:
            abscisses.append(i)
    if len(abscisses) %5 == 0 : break
```

```python
# Retrait des lignes en doubles quand le trait avait plus d'un pixel d'épaisseur
Abscisses = pd.DataFrame(abscisses, columns=['abscisses'])
Abscisses['abscisses_shifted'] = Abscisses.abscisses.shift(1)
Abscisses['diff'] = Abscisses.abscisses - Abscisses.abscisses_shifted
Abscisses = Abscisses[~(Abscisses['diff'] < 4)]
abscisses = list(Abscisses.abscisses)
```

```python
abscisses
```

```python
# Représentation horizontalement
plt.figure(figsize=(20,10))
plt.imshow(img_noir_blanc)
plt.title('Lignes de portée détectées', size=18)
for ab in abscisses:
    plt.axhline(y=ab,color='red')
plt.show()
```

Calcul de l'épaisseur d'une note unique. On adopte la convention que la tige d'une note de musique a une hauteur correspondant à 3 couloirs de notes, soit 3 quarts de portée.

```python
def get_taille_note(abscisses):
    diff = 0
    for i in range(len(abscisses)-1):
        diff += abscisses[i+1]-abscisses[i]
    largeur_note = int(diff / (len(abscisses)-1))
    hauteur_note = (max(abscisses) - min(abscisses))*(3/4)
    return largeur_note, hauteur_note

largeur_note, hauteur_note = get_taille_note(abscisses)
a = min(abscisses)-largeur_note
b = max(abscisses)+largeur_note
print('Une note de la partition prend en moyenne', largeur_note, 'pixels de large sur', hauteur_note, 'pixels de haut.')
```

### Création de la base de données
Comme la majorité des images ne contiennent pas de note de musique, on initialise les noms des notes avec "_0.png". Il faudra ensuite renommer les fichiers contenant une note (ex : "123_Mi.png", si l'image 123 contient un Mi).

```python
LARG, LONG = 14, 72 # dimensions choisies pour les images
```

```python
from matplotlib import patches
fig, ax = plt.subplots(figsize=(25,4))
ax.imshow(img_noir_blanc)
for i in np.arange(0, img_noir_blanc.shape[1], LARG):
    rect = patches.Rectangle((i,a), LARG, b-a, linewidth=1, edgecolor='r', facecolor='none')
    ax.add_patch(rect)
    plt.imsave('Datasets/'+str(i//LARG)+'_0.png', resize(img_noir_blanc[a:b, i:i+LARG], (LONG, LARG)))
plt.show()
```

Ces images vont être labelisées à la main, pour ensuite alimenter les dossiers Train et Valid.
